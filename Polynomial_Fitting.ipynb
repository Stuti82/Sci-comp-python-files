# Polynomial Fitting with numpy polyfit

Line: Polynomial of Degree 1

y = m*x +c

Quadratic equation : degree 2

y= a*x**2 + b*x + c

```python
import numpy as np
import matplotlib.pyplot as plt

x= np.linspace(-10,10,10)
y= 2*x + 5

plt.plot(x,y,'o')
```

# Two things can be done
## 1.  I have a set of data and I want to fit it to a polynomial

```python
polynomial_coefficients = np.polyfit(x,y,2)
```

```python
polynomial_coefficients
```

```python
xnew= np.linspace(-20,20,100)
ynew= np.poly1d(polynomial_coefficients)
plt.plot(xnew,ynew(xnew),x,y,'o')
```

# 2. Finding the value of the function at ANY point.

```python
ynew(0)
```

```python
ynew= y+ 5*np.random.rand(10)
plt.plot(x,ynew,'-o')
```

```python
pol_coeff=np.polyfit(x,ynew,2)
yfit= np.poly1d(pol_coeff)
plt.plot(xnew,yfit(xnew),x,ynew,'o')
```
